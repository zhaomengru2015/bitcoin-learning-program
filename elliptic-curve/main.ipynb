## 椭圆曲线
y<sup>2</sup> = x<sup>3</sup> + a*x+b

bitcoin secp256k1:

y<sup>2</sup> = x<sup>3</sup> + 7

### 特点


1. 椭圆曲线被一条直线横切，当这条直线是垂直，或者跟椭圆曲线正切，会有两个切点，否则的话，会有一个或者三个切点

2. 椭圆曲线上两个点的加法，就是找到这两个点的连线与椭圆曲线的第三个焦点，然后基于x轴做映射


### 加法满足以下特性：

1. Identity

   I+A=A

2. Commutativity

    A + B = B + A

3. Associativity

    (A + B) + C = A + (B + C)

4. 加法计算:

定义I为椭圆曲线上的无穷点：

a. 任何点与I点相加的结果还是I点

b. 垂直线上的两个点，相加的结果是I点

c. 两个不在垂直线上的不同点，也就是找出这两个点与椭圆曲线的第三个交点，满足：

P1 = (x1,y1), P2 = (x2,y2), P3 = (x3,y3) 

P1 +P2 =P3

s = (y2 – y1)/(x2 – x1)

x3 =s<sup>2</sup> –x1 –x2

y3 = s(x1 – x3) – y1

d. 两个不在垂直线上的相同点，也就是找出该点的切线与椭圆曲线的交点。满足：

P1 = (x1,y1), P3 = (x3,y3)

P1 +P1 =P3

s = (3*x1<sup>2</sup> + a)/(2y1)

x3 = s<sup>2</sup> – 2x1

y3 = s(x1 – x3) – y1

```python
class Point:
    def __init__(self,a,b,x,y):
        self.a=a
        self.b=b
        self.x=x
        self.y=y
        if self.x is None and self.y is None:
            return
        if pow(y,2)!=pow(x,3)+a*x+b:
            raise ValueError("init point failed")
    def __eq__(self,other):
        return True if self.y==other.y and self.x==other.x and self.a==other.a and self.b==other.b else False
    def __ne__(self,other):
        return True if self.y!=other.y or self.x!=other.x or self.a!=other.a or self.b!=other.b else False
    def __add__(self,other):
        if self.a!=other.a or self.b!=other.b:
            raise ValueError("a or b does not match")
        # (None,None) is I
        if self.x is None: 
            return other
        if other.x is None:
            return self
        # 一条直线与椭圆曲线有两个焦点，就是垂直线，相加就是I点。
        if self.x==other.x and self.y!=other.y:
            return self.__class__(self.a,self.b,None,None)
        # 一条直线与椭圆曲线只有一个交点，就是y值为0的点。相加也是I点。
        if self==other and self.y==0:
            return self.__class(self.a,self.b,None,None)
        # 一条直线与椭圆曲线有三个交点，但是这两个点是不同的点。(x3,y3)==(x1,y1)+(x2,y2)
        if self.x != other.x and self.y != other.y:
            s = int((other.y-self.y)/(other.x-self.x))
            x = int(pow(s,2) - self.x-other.x)
            y=int(s*(self.x-x)-self.y)
            return self.__class__(self.a,self.b,x,y)
        # 一条直线与椭圆曲线有三个交点，但是这两个点是相同的点。(x3,y3)=(x1,y1)+(x1,y1)
        s = int((3*pow(self.x,2) + self.a)/(2*self.y))
        x = pow(s,2)-2*self.x
        y = s*(self.x-x)-self.y
        return self.__class__(self.a,self.b,x,y)
    def __repr__(self):
        return f"Point_{self.x}_{self.y}_{self.a}_{self.b}"
# check point: y2 = x3 + 5x + 7
# p1=Point(5,7,2,4)
p2=Point(5,7,-1,-1)
p3=Point(5,7,18,77)
# p4=Point(5,7,5,7)
## Identity check
p1 = Point(5, 7,-1,-1)
p2 = Point(5, 7,-1, 1)
inf = Point(5, 7,None,None)
print("p1+inf=p1: {}".format(p1+inf==p1))
print("p2+inf=p2: {}".format(p2+inf==p2))
p1 = Point(5, 7,2,5)
p2 = Point(5, 7,-1,-1)
print("p2+p2=: {}".format(p1+p2))
p1 = Point(5, 7,-1,-1)
p2 = Point(5, 7,-1, 1)
print("p1+p2=inf: {}".format(p1+p2))
p1=Point(5, 7,-1,-1)
p2=Point(5, 7,-1,-1)
print("p1+p2={}".format(p1+p2))
```
